```r
setwd("~/R/code-base/analyticvidya/titanic/")
train = read.csv("train.csv")
test = read.csv('test.csv')
```

```r
# let's use decision tree 
# recursive partitioning and regression trees (rpart)
library(rpart)
```

```r
# method = 'class' if binary outcome
# method = 'anova' if continuous outcome
fit = rpart(Survived ~ Pclass + Sex + Age + SibSp + Parch + Fare + Embarked,
            data = train, method = 'class')
```

```r
# to examine create tree 
plot(fit)
```

```r
plot(fit)
text(fit)
```

```r
# To get some more information graphics
library(rattle)
library(rpart.plot)
library(RColorBrewer)
```

```r
 fancyRpartPlot(fit)
```

```r
prediction = predict(fit, test, type ='class')
prediction
```

```r
submit <- data.frame(PassengerId = test$PassengerId, Survived = prediction)
write.csv(submit, file = "myfirstdtree.csv", row.names = FALSE)
```

```r
# let's optimize some of the default parameter of rpart
#cp - metric that stop splits that are not deemed enough
# minsplit = metric  that determine how many data should be there in a bucket before split
# all these are menupulated by control

fit = rpart(Survived ~ Pclass + Sex + Age + SibSp + Parch + Fare + Embarked,
           data = train, method = 'class', control = rpart.control(cp=0, minsplit =2))
fancyRpartPlot(fit)
```

```r
# but this leads to overfitting let's submit again
prediction = predict(fit, test, type='class')
```

```r
submit = data.frame(PassengerID=test$PassengerId, Survived = prediction)
write.csv(submit, file='mymodifieddr.csv', row.names=FALSE)
```

```r
# model accuracy doesn't come better for test data due to overfitting
```

```r
 fit = rpart(Survived ~ Pclass + Sex + Age + SibSp + Parch + Fare + Embarked,
               data=train,
               method="class",
               control=rpart.control(cp=0.02, minsplit = 2))
new.fit <- prp(fit,snip=TRUE)$obj
fancyRpartPlot(new.fit)
```
